# To fit a continumm to the flux data

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# importing the data
pix, fl, fle = np.loadtxt('sky_sub_2009oct30_0009.fits_flux.dat', usecols=(0,1,2), unpack=True)
# Let's look at the spectrum
plt.figure(figsize=(16*1.5, 9*1.5))
plt.errorbar(pix, fl, yerr=fle)
plt.grid()
```

```python
# Let's now fit linear polynomial to it
aa = np.polyfit(pix, fl, deg=1, w=fle, cov=True)
```

```python
print(aa[0])
print(aa[1])
```

```python
def line(x, a0, a1):
    return a1*x+ a0

def qua(x, a0, a1, a2):
    return a2*x**2 + a1*x + a0
```

```python
plt.figure(figsize=(16*1.5,9*1.5))
plt.errorbar(pix, fl, yerr=fle)
plt.plot(pix, line(pix, aa[0][1], aa[0][0]))
```

```python
# Now let's fit a quadratic model
bb = np.polyfit(pix, fl, deg=2, w=fle, cov=True)
```

```python
print(bb[0])
print(bb[1])
```

```python
plt.figure(figsize=(16*1.5,9*1.5))
plt.errorbar(pix, fl, yerr=fle)
plt.plot(pix, qua(pix, bb[0][2], bb[0][1], bb[0][0]))
```

```python
def fitting(datafile, deg):
    x, y, ye = np.loadtxt(datafile, usecols=(0,1,2), unpack=True)
    pp = np.polyfit(x, y, w=ye, deg=deg)
    return pp

data = 'sky_sub_2009oct30_0009.fits_flux.dat'
ab = fitting(data, 3)

plt.figure(figsize=(16*1.5,9*1.5))
plt.errorbar(pix, fl, yerr=fle)
plt.plot(pix, ab[0]*pix**3 + ab[1]*pix**2 + ab[2]*pix + ab[3])
```

```python
aaa = np.array([1,2,3])
print(np.flip(aaa))
```

```python
def arbi_poly(x, *params):
    return sum([p*(x**i) for i, p in enumerate(params)])

abc = np.flip(fitting(data, 10))
penta = arbi_poly(pix, *abc)

plt.figure(figsize=(16*1.5,9*1.5))
plt.errorbar(pix, fl, yerr=fle)
plt.plot(pix, penta)
```

```python
def fit_poly(datafile, deg, plot=True):
    x, y, ye = np.loadtxt(datafile, usecols=(0,1,2), unpack=True)
    ab1 = np.polyfit(x, y, w=ye, deg=deg)
    ab2 = np.flip(ab1)
    polynom = arbi_poly(x, *ab2)
    if plot==True:
        plt.errorbar(x, y, yerr=ye)
        plt.plot(x, polynom)
    return ab1
```

```python
def fit_poly_norm(datafile, deg, selection, plot=True):
    x, y, ye = np.loadtxt(datafile, usecols=(0,1,2), unpack=True)
    we = np.copy(ye)
    for i in range(len(selection)):
        sel = selection[i].split(':')
        we[int(sel[0]):int(sel[1])] = 0
    ab1 = np.polyfit(x, y, w=we, deg=deg)
    ab2 = np.flip(ab1)
    polynom = arbi_poly(x, *ab2)
    if plot==True:
        plt.errorbar(x, y, yerr=ye)
        plt.plot(x, polynom)
    return ab1
```

```python
def normal_spectrum(datafile, degree, selection):
    x, y, ye = np.loadtxt(datafile, usecols=(0,1,2), unpack=True)
    we = np.copy(ye)
    for i in range(len(selection)):
        sel = selection[i].split(':')
        we[int(sel[0]):int(sel[1])] = 0
    ab1 = np.polyfit(x, y, w=we, deg=degree)
    ab2 = np.flip(ab1)
    polynom = arbi_poly(x, *ab2)
    fl_norm = y/polynom
    fle_norm = ye/polynom
    return x, fl_norm, fle_norm
```

```python
data_files = ['sky_sub_2009oct30_0009.fits_flux.dat', 'sky_sub_2009oct30_0010.fits_flux.dat', 'sky_sub_2009oct30_0011.fits_flux.dat']
selec = [['1:13', '135:142', '328:360', '413:464', '508:519', '727:794'], \
         ['81:130', '326:363', '730:780', '866:876', '888:894'], \
         ['30:54', '310:362', '410:465', '510:545', '697:717', '726:780']]
plt.figure(figsize=(16*1.5, 9*1.5))
for i in range(len(data_files)):
    nor_pix, nor_fl, nor_fle = normal_spectrum(data_files[i], degree=3, selection=selec[i])
    plt.errorbar(nor_pix, nor_fl, yerr=nor_fle, label='Observation No. ' + str(i))

plt.grid()
plt.legend(loc='best')
```
